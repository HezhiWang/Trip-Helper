```python
def distance(lat1,lng1,lat2,lng2):
    add1 = (lat1,lng1)
    add2 = (lat2,lng2)
    d = vincenty(add1, add2).miles
    return d
```

```python
def within_distance(df, centerlat, centerlng, distance_within):
    '''find index of locations in dataframe df 
    that's within a given distance of a given center'''
    ind = []
    for i in range(0,df.shape[0]):
        lat = df['Lat'].ix[i]
        lng = df['Lng'].ix[i]
        if distance(lat,lng,centerlat,centerlng) <= distance_within:
            ind.append(i)
    return ind
```

```python
def sort_locations(df, given_filter, value):
    '''find locations in df with the given value of the given filter, 
    and then sort by score and total reviews'''
    df_sub =df[df[given_filter]==value]
    df_sorted = df_sub.sort_values(by = ['Avgscore','Total_review'],ascending=[False, False])
    return df_sorted
```

```python
ind = within_distance(data, 40.748817, -73.985428, 1)
df_sorted = sort_locations(data.ix[ind], 'Price', 3)
```
